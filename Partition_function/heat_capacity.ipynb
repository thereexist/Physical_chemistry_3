## Heat capacity of $H_2O$

중간 레포트 4번 문제에 대한 문제 해설입니다.

### 1. Partition function of Translational motion & Portion of Translational motion in $C_V$(isochoric heat capacity).

Single particle에 대해 translational motion으로 기술되는 partition function은 다음과 같습니다.

$$q_{trans}(V,T) = \frac{V}{\Lambda(T)^3}$$

$$\Lambda(T) = \frac{h}{\sqrt{2 \pi m k_B T}}$$

이를 이용해 single particle의 translational motion이 가지고 있는 energy는 다음과 같이 구할 수 있습니다.

$$\braket{\varepsilon^{trans}} = -(\frac{\partial \ln q_{trans}}{\partial \beta}) = \frac{\partial T}{\partial \beta} \frac{\partial \ln \Lambda(T)^3}{\partial T} = -\frac{1}{k_B \beta^2} \times -\frac{3}{2} \frac{\partial \ln T}{\partial T} = \frac{3}{2k_B \beta^2}  \times \frac{1}{T} = \frac{3}{2} k_B T$$



그렇다면, N개의 particle로 구성된 system에서 translational motion(병진 운동, 직선 운동)이 가지게 되는 energy는 $U=\braket{\varepsilon^{trans}} = 3 N k_B T / 2$로 쓸 수 있게 됩니다.! (어디서 많이 본 식의 형태죠?)

N개의 particle의 translational motion에 의해 발생하는 heat capacity는 다음과 같이 기술할 수 있습니다.


$$(C_{V})^{trans}=N \times \frac{\partial\braket{\varepsilon^{trans}}}{\partial T} = \frac{3}{2} N k_B$$

### 2. Partition function of rotational motion & Portion of rotational motion in $C_V$(isochoric heat capacity).

위에서 계산했던 것과 같이 계산을 해봅시다. 여기에서 $H_2 O$를 계산한다는 것을 감안해, asymmetric top의 partition function을 사용하면 다음과 같은 결과들을 얻을 수 있습니다. 

$$q_{rot} = \frac{\pi^{1/2}}{\sigma} (\ \frac{T^3}{\Theta_{rot,A} \Theta_{rot,B} \Theta_{rot,C}})\ ^ {1/2} \text{   for asymmetric top}$$

$$\braket{\varepsilon^{rot}} = -(\frac{\partial \ln q_{rot}}{\partial \beta}) = -\frac{3}{2}\frac{\partial T}{\partial \beta} \frac{\partial \ln T}{\partial T} = \frac{3}{2} \frac{1}{k_B \beta^2} \times \frac{1}{T} = \frac{3}{2} \frac{1}{k_B \beta^2}  \times \frac{1}{T} = \frac{3}{2} k_B T$$




$$(C_{V})^{rot}=N \times \frac{\partial\braket{\varepsilon^{rot}}}{\partial T} = \frac{3}{2} Nk_B$$

### 3. Partition function of vibrational motion & Portion of vibrational motion in $C_V$(isochoric heat capacity).

마지막으로, vibrational motion에 대해 계산을 해볼까요? 계산의 결과는 아래와 같습니다.

우선은 단 한개의 $\nu_i$만 고려해서 계산을 진행했습니다.



$$q_{vib,i}(T) = \frac{e^{-\beta h \nu_i /2}}{1-e^{-\beta h \nu_i}}$$


$$\braket{\varepsilon^{vib}}_i = -(\frac{\partial \ln q_{vib,i}}{\partial \beta}) = \frac{h \nu_i}{2} + \frac{h \nu_i e^{-\beta h \nu_i}}{1 - e^{-\beta h \nu_i}}$$


$$(C_{V})_i^{vib} = N \times \frac{\partial\braket{\varepsilon^{vib}}_i}{\partial T} = N \times \frac{\partial \beta}{\partial T} \frac{\partial\braket{\varepsilon^{vib}}_i}{\partial \beta} = -N k_B \beta^2 \times -\frac{h^2 \nu_i^2 e^{-\beta h \nu_i}}{(1 - e^{-\beta h \nu_i})^2} = N k_B \frac{(\beta h \nu_i)^2 e^{-\beta h \nu_i}}{(1 - e^{-\beta h \nu_i})^2} $$

$(C_{V})^{vib}$를 제대로 계산을 하기 위해서는 각각의 vibration mode를 모두 고려해주어야 됩니다. 

수업시간에 배웠던 것을 생각해본다면, $H_2O$의 경우, 3개의 vibration mode가 있겠죠?

이를 모두 고려해주면 다음과 같은 결과를 얻을 수 있습니다.

$$(C_{V})^{vib} = \sum_{i=1}^3 (C_{V})_i^{vib} = N k_B \sum_{i=1}^3 \frac{(\beta h \nu_i)^2 e^{-\beta h \nu_i}}{(1 - e^{-\beta h \nu_i})^2} = nR \sum_{i=1}^3 \frac{(\Theta_i / T)^2 e^{-\Theta_i / T}}{(1 - e^{-\Theta_i / T})^2} $$

위의 translational motion과 rotational motion과 달리, vibrational motion의 heat capacity는 상수로 나타나지 않는다는 것을 알 수 있습니다.

$(\Theta_i / T)^2 e^{-\Theta_i / T}/ (1 - e^{-\Theta_i / T})^2 $ 값에 따라 vibration이 heat capacity에 기여하는 정도가 달라지게 됩니다.

$T$의 함수라는 것은 온도에 따라 기여하는 정도가 달라진다는 것인데,,,, 이를 아래의 python code를 통해 확인해봅시다.

```python
# import libraries
%matplotlib inline
from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np
```

```python
# function definition
def vib(Θ, T):
    return (Θ/T)**2 * np.exp(-Θ/T) / (1 - np.exp(-Θ/T))**2


# interactive_plot function : Temperature range와 Θ_vib 값을 직접 조절하면서 graph를 그려보세요.
def vib_plot(T_max = 1000, Θ_1 = 5360, Θ_2 = 5160, Θ_3 = 2290):
    plt.figure(2)
    T_min = 10
    T_array = np.linspace(T_min,T_max, 1000)

    plt.figure(figsize = (8,6))
    plt.plot(T_array, vib(Θ_1, T_array), 'r', label = "Θ_1", lw = 4)
    plt.plot(T_array, vib(Θ_2, T_array), 'b', label = "Θ_2", lw = 4)
    plt.plot(T_array, vib(Θ_3, T_array), 'g', label = "Θ_3", lw = 4)
    plt.xlabel("Temperature (K)",fontsize = 15)
    plt.ylabel(r"$(\Theta_i / T)^2 e^{-\Theta_i / T}/ (1 - e^{-\Theta_i / T})^2$",fontsize = 15)
    plt.xticks(fontsize = 15)
    plt.yticks(fontsize = 15)
    plt.legend(fontsize = 15)
    plt.show()
```

```python
interactive_plot = interactive(vib_plot, T_max = (100, 20000, 10), Θ_1 = (0, 8000, 10), Θ_2 = (0, 8000, 10), Θ_3 = (0, 8000, 10))
output = interactive_plot.children[-1]
output.layout.height = '480px'
interactive_plot
```

```python
vib_plot(T_max = 1000, Θ_1 = 5360, Θ_2 = 5160, Θ_3 = 2290)
```

위에서 구한 결과들을 이용해서 물의 $C_V / (N k_B)$를 구해보고, 이를 실험값과 비교해봅시다.

실험의 결과는 $C_P$ 측정 결과로 ([**출처**](http://www.ddbst.com/en/EED/PCP/ICP_C174.php)), $C_P = C_V + nR$의 관계식을 이용해 $C_V$로 변환해서 비교했습니다.

관측하고자 하는 물질에 직접 열을 가하면서(매우 천천히) 측정을 하게 됩니다.

위의 process는 액체의 heat capacity를 측정하는 내용이지만, 기체도 특별히 다르지는 않을 것입니다.(직접 측정을 해본적은 없어서 100% 확신은 없네요..)

```python
import pandas as pd

# export experiment data
experiment_data = pd.read_csv("./Heat_capacity_H2O.csv")

# define function
def R_over_molar_heat_capacity(Θ_1, Θ_2, Θ_3, T):
    return 3 + vib(Θ_1, T) + vib(Θ_2, T) + vib(Θ_3, T)

T_min = 10
T_max = 6000
T_array = np.linspace(T_min,T_max, 1000)

Θ_1 = 5360
Θ_2 = 5160
Θ_3 = 2290

plt.figure(figsize = (8,6))
plt.plot(T_array, R_over_molar_heat_capacity(Θ_1, Θ_2, Θ_3, T_array), 'k', label = "Theoretical prediction", lw = 4)
plt.plot(experiment_data['T [K]'], experiment_data['C_P / (N * k_B)'] - 1, 'ro', markersize = 3, label = "Experimental data") # experiment data를 가지고 왔습니다.
plt.plot([-10000,10000],[6,6], "k--",label = "6", lw = 2)
plt.xlim(-100,6100)
plt.xlabel("Temperature (K)",fontsize = 15)
plt.ylabel(r"$C_V / (N k_B)$",fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

결과를 확인해보니, 제 예상과 다른 결과가 나왔습니다. 여러분들은 이런 질문을 하실 수 있게 되어버린거죠.

## New data (NIST data)

NIST의 shomate equation([**출처**](https://webbook.nist.gov/cgi/cbook.cgi?ID=C7732185&Mask=1&Type=JANAFG&Plot=on#JANAFG))를 이용해서 이론과 비교를 해봤습니다.

```python
T_array2 = np.linspace(1700,6000) # Chase, 1998 temperature range
shomate_array2 = np.zeros_like(T_array2)

# Chase, 1998 Shomate equation
A = 41.96426
B = 8.622053
C = -1.499780	
D = 0.098119	
E = -11.15764	
F = -272.1797	
G = 219.7809	
H = -241.8264	

def shomate_value(T):
    t = T / 1000
    Cp0 =  A + B*t + C*t**2 + D*t**3 + E/t**2 
    R = 8.3145 # [J / (K mol) = N m / (K mol) = kg (m/s)**2 / (K mol)]

    return Cp0 / R - 1
    
count = 0
for T in T_array2:
    shomate_array2[count] = shomate_value(T)
    count += 1

plt.figure(figsize = (8,6))
plt.plot(T_array, R_over_molar_heat_capacity(Θ_1, Θ_2, Θ_3, T_array), 'k', label = "Theoretical prediction", lw = 4)
plt.plot(experiment_data['T [K]'], experiment_data['C_P / (N * k_B)'] - 1, 'ro', markersize = 3, label = "Experimental data") # experiment data를 가지고 왔습니다.
plt.plot(T_array2, shomate_array2, 'g--', label = "Shomate equation", lw = 4)
plt.plot([-10000,10000],[6,6], "k--",label = "6", lw = 2)
plt.xlim(0,6000)
plt.xlabel("Temperature (K)",fontsize = 15)
plt.ylabel(r"$C_V / (N k_B)$",fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15)
plt.show()
```

## 추가 내용(위의 문제 해결됨.)

[**1956년에 나온 논문**](https://nvlpubs.nist.gov/nistpubs/jres/56/jresv56n2p105_A1b.pdf)에는 translational motion을 제외한 rotation과 vibration에 의한 partition function이 엄밀하게 계산되어 있습니다.

rotation에 대해서는 저희가 배운 rotational partition function뿐 아니라 rotational correction factor가 추가 되었으며,

rotation과 vibration의 correlation하면서 발생하는 partition function이 추가 되었습니다.

--------------------------------------------------------------------------------------------------------

rotational partition function과 rotational correction factor는 [**1954년에 나온 논문**](https://aip.scitation.org/doi/abs/10.1063/1.1739991)에 자세하게 기술되어 있습니다.

($\frac{\sqrt{\pi}}{\sigma} \left( \frac{T^3}{\Theta_{rot,A} \Theta_{rot,B} \Theta_{rot,C}} \right) ^ {1/2}$을 저희가 계산을 직접한 기억이 있나요? $\frac{\sqrt{\pi}}{\sigma} \left( \frac{T}{\Theta_{rot}} \right) ^{3/2}$은 직접 계산 했어도 Asymmetric top의 rotational partition function을 직접 계산한 기억은 없죠?)

엄밀한 계산을 통해 얻은 Asymmetric top의 rotational partition function은 다음과 같습니다.

$$q_{rot} \cong \frac{\pi^{1/2}}{\sigma} \left( \frac{T^3}{\Theta_{rot,A} \Theta_{rot,B} \Theta_{rot,C}} \right) ^ {1/2} \times \left( 1 + \frac{\theta_1}{T} + \frac{\theta_2}{T^2} \right) \times (1 + \rho T)$$

여기에서 $\rho$는 stretching constant로, 온도에 따라 증가하는 bond length에 의해 moment of Inertia가 변하는 것을 고려한 값입니다.(정확한 derivation은 [**여기**](https://aip.scitation.org/doi/pdf/10.1063/1.1749895)에 나옵니다.)

그리고 $\left( 1 + \frac{\theta_1}{T} + \frac{\theta_2}{T^2} \right)$ 는 low temperature quantum correction for rotation이 고려된 term입니다.

이 rotational correction factor $\left( = \left( 1 + \frac{\theta_1}{T} + \frac{\theta_2}{T^2} \right) \times (1 + \rho T) \right)$ 는 low temperature에서는 heat capacity에 영향을 미치지만, high temperature에서는 영향을 미치지 못하는 term입니다.

---------------------------------------------------------------------------------------------------------------

rotation과 vibration의 correlation term은 [**1961년에 나온 논문**](https://aip.scitation.org/doi/abs/10.1063/1.1732232)에 approximately하게 다음과 같이 기술되어 있습니다.

$$\ln q_{correlation} \cong C_0(T) + C_1(T)/T + C_2(T)/T^2$$

이 term은 high temperature에서 heat capacity에 크게 영향을 미치는 term입니다.

[**1961년에 나온 논문**](https://aip.scitation.org/doi/abs/10.1063/1.1732232)의 Table 1을 보면 온도에 따른 contribution이 자세하게 나와 있습니다.

이 효과로 인해 high temperature에서 $C_V / (N k_B)$가 6에 수렴하는 것이 아닌, 그보다 큰 값으로 나타나게 됩니다.

------------------------------------------------------------------------------------------------------------------


아.. 이것이 의미하는 바는 무엇일까요? 저희가 이전에 배웠던 [**"Equipartition theorem"**](https://www.wikiwand.com/en/Equipartition_theorem)에 rotation-vibration correlation으로 인한 degree of freedom이 추가된다고 볼 수 있을 것 같습니다.

특히 매우 고온에서 이것이 발현된다고 볼 수 있습니다.
