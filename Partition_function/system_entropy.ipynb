## System entropy

[![Hits](https://hits.seeyoufarm.com/api/count/incr/badge.svg?url=https%3A%2F%2Fgithub.com%2Fthereexist%2FPhysical_chemistry_3%2Fblob%2Fmain%2FPartition_function%2Fsystem_entropy.ipynb&count_bg=%2379C83D&title_bg=%23555555&icon=&icon_color=%23E7E7E7&title=hits&edge_flat=false)](https://hits.seeyoufarm.com)

system entropy는 다음과 같이 계산을 할 수 있습니다.

$$ S = (U-A)/T = -\frac{1}{T}(\partial \ln Q / \partial \beta)_{N,V} + k_B \ln Q = k_B T (\partial \ln Q / \partial T)_{N,V} + k_B \ln Q$$

이때 첫번째 term과 두번쨰 term을 나눠서 계산을 진행해보겠습니다.

### A) 첫번째 term

$k_B T (\partial \ln Q / \partial T)_{N,V}$ 에서 $(\partial \ln Q / \partial T)_{N,V}$ 부터 계산을 해보도록 하겠습니다.

$$ Q = \frac{q^N}{N!}$$

$$\partial \ln Q / \partial T = N (\partial \ln q / \partial T)$$

$$ \partial \ln q / \partial T = \partial \ln q_{trans} / \partial T + \partial \ln q_{rot} / \partial T + \partial \ln q_{vib} / \partial T + \partial \ln q_{elec} / \partial T $$ 


4개의 partition function term을 고려해 식을 다시 쓰면 다음과 같이 쓸 수 있습니다. 

$$ \partial \ln q / \partial T = \partial \left( \ln \left( V / \Lambda(T)^3 \right) \right) / \partial T + \partial  \ln q_{rot}  / \partial T + \partial \left( \ln \left( \prod_{j=1}^{j_{max}} \frac{1}{(1-e^{-\Theta_{vib,j}/T})} \right) \right) / \partial T$$ 

이제 위의 식에서 나온 3개의 term을 또 다시 각각 계산을 해보겠습니다.

1)

$$ \frac{\partial}{\partial T} \left( \ln \frac{V}{\Lambda(T)^3} \right) = -3 \partial \ln \Lambda(T) / \partial T = \frac{3}{2} \frac{\partial}{\partial T} \left(\ln T \right) = \frac{3}{2T}$$

$$(\because \Lambda(T) \thicksim T^{-1/2})$$



2)

$$ \partial \ln q_{rot} / \partial T = \frac{f_{rot}}{2T}$$

3)

$$ \partial \left( \ln \left( \prod_{j=1}^{j_{max}} \frac{1}{(1-e^{-\Theta_{vib,j}/T})} \right) \right) / \partial T= -\partial \left( \sum_{j=1}^{j_{max}} \ln (1-e^{-\Theta_{vib,j}/T}) \right) / \partial T$$



$$ -\partial \left( \sum_{j=1}^{j_{max}} \ln (1-e^{-\Theta_{vib,j}/T}) \right) / \partial T = -\sum_{j=1}^{j_{max}} \left( \frac{-\Theta_{vib,j} T^{-2}e^{-\Theta_{vib,j}}}{1 - e^{-\Theta_{vib,j}/T}} \right) = \frac{1}{T^2} \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j}}{e^{\Theta_{vib,j}/T}-1} \right)$$

위에서 구한 결과들을 취합하면 다음과 같은 결과를 얻을 수 있습니다.

$$\therefore  \partial \ln q / \partial T  = \frac{3 + f_{rot}}{2T} + \frac{1}{T^2} \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j}}{e^{\Theta_{vib,j}/T}-1} \right)$$

$$ \therefore k_B T (\partial \ln Q / \partial T)_{N,V} = N k_B \frac{3 + f_{rot}}{2} + N k_B \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j} / T}{e^{\Theta_{vib,j}/T}-1} \right)$$

잘 따라오셨나요? 이제 두번쨰 term을 한번 풀어보겠습니다.

### B) 두번쨰 term

$$\ln Q = N \ln q - \ln N! \cong N \ln q - N \ln N + N = N \ln \frac{V}{\Lambda(T)^3} + N \ln q_{rot} + N \ln \left( \prod_{j=1}^{j_{max}} \frac{1}{(1-e^{-\Theta_{vib,j}/T})} \right) + N \ln g_{el} - N \ln N + N$$

$$\therefore k_B \ln Q = N k_B \left( \ln \frac{V}{\Lambda(T)^3} + \ln q_{rot} + \ln \left( \prod_{j=1}^{j_{max}} \frac{1}{(1-e^{-\Theta_{vib,j}/T})} \right) + \ln g_{el} - \ln N + 1 \right)$$

위에서 구한 첫번쨰 term과 두번쨰 term을 더하고 정리를 하면 entropy에 대한 식을 다음과 같이 정리할 수 있습니다.

$$\therefore \frac{S}{N k_B} = \frac{T}{N} \left( \partial \ln Q / \partial T \right)_{N,V} + \frac{1}{N} \ln Q = \frac{3 + f_{rot}}{2} + \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j} / T}{e^{\Theta_{vib,j}/T}-1} \right) + \left( \ln \frac{V}{\Lambda(T)^3} + \ln q_{rot} + \ln \left( \prod_{j=1}^{j_{max}} \frac{1}{(1-e^{-\Theta_{vib,j}/T})} \right) + \ln g_{el} - \ln N + 1 \right)$$

$$=\left( \frac{5}{2} + \ln \frac{V}{N\Lambda(T)^3} \right) + \left( \ln q_{rot} + \frac{f_{rot}}{2}\right) + \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j} / T}{e^{\Theta_{vib,j}/T}-1} - \ln \left( 1-e^{-\Theta_{vib,j}/T} \right)\right) + \ln g_{el}$$

$$=\left( \frac{5}{2} + \ln \frac{1}{\rho N_{Avo}\Lambda(T)^3} \right) + \left( \ln q_{rot} + \frac{f_{rot}}{2}\right) + \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j} / T}{e^{\Theta_{vib,j}/T}-1} - \ln \left( 1-e^{-\Theta_{vib,j}/T} \right)\right) + \ln g_{el}$$

$$\left( \rho \equiv \frac{n}{V} = \frac{N}{N_{Avo} V} \right)$$

System entropy에 대해 이론적인 결과를 위와 같이 얻었습니다.

이 값은 실험결과와 얼마나 일치할까요?

실험결과와 비교를 하기 위해서는 "1 mol 당 entropy" 라는 정의를 가진 "Standard molar entropy"와 비교를 하는 것이 용이합니다.

위에서 구한 결과를 이용해 정리를 하면 standard molar entropy는 다음과 같이 적을 수 있습니다.

$$\bar{S}^{\circ} \equiv S / n =R \left( \frac{5}{2} + \ln \frac{1}{\rho N_{Avo}\Lambda(T)^3} \right) + R \left( \ln q_{rot} + \frac{f_{rot}}{2}\right) + R \sum_{j=1}^{j_{max}} \left( \frac{\Theta_{vib,j} / T}{e^{\Theta_{vib,j}/T}-1} - \ln \left( 1-e^{-\Theta_{vib,j}/T} \right)\right) + R \ln g_{el}$$

자 이제, PDF 마지막 2 페이지에 걸쳐 나와있는 thermochemical tables를 이용해서 $N_2O$와 $NO_2$의 standard molar entropy를 계산해봅시다.

```python
import numpy as np

T = 298.15 # [K]
P = 10**5  # [N m**-2]

R = 8.3145 # [J / (K mol) = N m / (K mol) = kg (m/s)**2 / (K mol)]

N_avo = 6.022 * 10**23 # [dimensionless]

k_B = 1.38 * 10**-23 # [J/K = N m / K = kg m/s**2 m/K = kg (m/s)**2 / K]

h = 6.626*10**-34   # [m**2 kg/s]

def ρ(T):
    return P / (R * T) # [mol / m**3]

def mass(M_r): # molecular weight to mass per particle
    return M_r / N_avo * 10**-3 # [kg]

def Λ(m,T):
    return h / np.sqrt(2*np.pi*m*k_B*T) # [m]

def Θ_vib(ν_tilde): # ν_tilde : [cm**-1]
    return (6.626 * 10**-34)*(3.0*10**10) / (1.380 * 10**-23) * ν_tilde # [K]

def S_bar_vib(Θ_vib_j,T):
    value = Θ_vib_j / T
    return R * (value / (np.exp(value) - 1) - np.log(1 - np.exp(-value))) # [J / (K mol)]

def Θ_rot(B_tilde):
    return (6.626 * 10**-34)*(3.0*10**10) / (1.380 * 10**-23) * B_tilde # [K]

def q_rot_rod(Θ_rot, T):
    # rotational partition function for rod shape molecule: N2O

    return T / Θ_rot

def q_rot_top(product_of_Inertia,σ, T):
    # rotational partition function for top shape molecule: NO2
    product_of_rot_temp = (h**2 / (8*np.pi**2*k_B))**3 * (product_of_Inertia*10**-21)**-1 # 10**-21 for g**3 cm**6 to kg**3 * m**6
                            # product_of_rot_temp : [K**3]
    return np.sqrt(np.pi) / σ * (T**3 / product_of_rot_temp)**(1/2) 
```

```python
## N2O standard molar entropy

T = 298.15 # [K]

f_rot_N2O = 2
g_el_N2O = 1

result_N2O = R * (5/2 - np.log(ρ(T) * N_avo * Λ(mass(M_r = 44.0128),T = T)**3)) 
result_N2O += R * (np.log(q_rot_rod(Θ_rot(B_tilde = 0.421019),T = T)) + f_rot_N2O/2)
result_N2O += (S_bar_vib(Θ_vib(ν_tilde = 1276.5),T = T) + S_bar_vib(Θ_vib(ν_tilde = 589.2),T = T) * 2 + S_bar_vib(Θ_vib(ν_tilde = 2223.7),T = T))
result_N2O += R * np.log(g_el_N2O)

print(f"Theoretical prediction of \"standard molar entropy of N2O\" is {result_N2O} [J / (K mol)].")
print("Experimental result of \"standard molar entropy of N2O\" is 219.957 [J / (K mol)]")

error = np.abs(result_N2O - 219.957) / 219.957 * 100 # [%]

print(f"Error = {error} %")
```

```python
## NO2 standard molar entropy

T = 298.15 # [K]

f_rot_NO2 = 3
g_el_NO2 = 2
result_NO2 = R * (5/2 - np.log(ρ(T) * N_avo * Λ(mass(M_r = 46.0055),T = T)**3)) 
result_NO2 += R * (np.log(q_rot_top(product_of_Inertia = 1.54229 * 10**-116, σ = 2, T = T)) + f_rot_NO2/2)
result_NO2 += (S_bar_vib(Θ_vib(ν_tilde = 1357.8),T = T) + S_bar_vib(Θ_vib(ν_tilde = 756.8),T = T) + S_bar_vib(Θ_vib(ν_tilde = 1665.5),T = T))
result_NO2 += R * np.log(g_el_NO2)

print(f"Theoretical prediction of \"standard molar entropy of NO2\" is {result_NO2}[J / (K mol)].")
print("Experimental result of \"standard molar entropy of NO2\" is 240.034 ± 0.13 [J / (K mol)]")

error = np.abs(result_NO2 - 240.034) / 240.034 * 100 # [%]

print(f"Error = {error} %")
```

결과를 확인해보면, 실험 결과와 이론 결과가 놀라울 정도로 잘 맞는걸 확인할 수 있습니다.

그렇다면 위의 standard molar entropy는 이론적으로 특정 온도에서만 실험값과 일치하는 것일까요?(우연히?, 과연?)

### 이를 확인해봅시다.

[**출처**](https://webbook.nist.gov/cgi/cbook.cgi?ID=C10024972&Mask=1#Thermo-Gas)를 보시면, $N_2O$의 Standard entropy에 대해 Shomate Equation이 존재한다는 사실을 확인할 수 있습니다.

Shomate equation이란, 실험 데이터를 fitting한 결과 함수라고 생각하면 됩니다.

이 Shomate equation은 두가지 문제를 가지고 있습니다.

1. 실험 데이터의 결과를 맞추는 함수이지만, 그 함수가 생긴 근본은 경험에 근거한 empirical equation이기에, "왜 그런 결과가 나오는가?" 라는 질문에 대해서는 답을 해주기 어려운 식입니다.

2. 또한 보시면 temperature region이 낮은 온도(lower temperature)와 높은 온도(higher temperature)인가에 따라 식이 두개로 나뉘어져있다는 것을 보실 수 있습니다. 한번에 모든 온도에 대해 설명을 하지 못하는 식이라는 것을 의미합니다.

과연 저희가 구한 이론결과는 이와 잘 맞을지, 확인해봅시다.

```python
import matplotlib.pyplot as plt

## N2O standard molar entropy 1. Lower Temperature

T_array = np.linspace(298,1400) # Chase, 1998 temperature range
theoretical_array = np.zeros_like(T_array)
shomate_array = np.zeros_like(T_array)

# Theoretical value
f_rot_N2O = 2
g_el_N2O = 1
count = 0
for T in T_array:
    result_N2O = R * (5/2 - np.log(ρ(T) * N_avo * Λ(mass(M_r = 44.0128),T)**3)) 
    result_N2O += R * (np.log(q_rot_rod(Θ_rot(B_tilde = 0.421019),T)) + f_rot_N2O/2)
    result_N2O += (S_bar_vib(Θ_vib(ν_tilde = 1276.5),T) + S_bar_vib(Θ_vib(ν_tilde = 589.2),T) * 2 + S_bar_vib(Θ_vib(ν_tilde = 2223.7),T))
    result_N2O += R * np.log(g_el_N2O)
    theoretical_array[count] = result_N2O
    count += 1

# Chase, 1998 Shomate equation
A = 27.67988
B = 51.14898
C = -30.64454
D = 6.847911
E = -0.157906
F = 71.24934
G = 238.6164
H = 82.04824

count = 0
for T in T_array:
    t = T / 1000
    value = A*np.log(t) + B*t + C*t**2/2 + D*t**3/3 - E /(2*t**2) + G
    shomate_array[count] = value
    count += 1


plt.figure(figsize = (8,6))
plt.plot(T_array, theoretical_array, 'r-', lw = 4, label = "Theory")
plt.plot(T_array, shomate_array, 'g--', lw = 4, label = "Shomate equation, chase 1988")
plt.legend()
plt.xlabel("Temperature (K)", fontsize = 15)
plt.ylabel("Standard molar entropy of N2O $( J / (K mol))$", fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15)
```

```python
## N2O standard molar entropy 2. Higher temperature

T_array2 = np.linspace(1400,6000) # Chase, 1998 temperature range
theoretical_array2 = np.zeros_like(T_array2)
shomate_array2 = np.zeros_like(T_array2)

# Theoretical value
f_rot_N2O = 2
g_el_N2O = 1
count = 0
for T in T_array2:
    result_N2O = R * (5/2 - np.log(ρ(T) * N_avo * Λ(mass(M_r = 44.0128),T)**3)) 
    result_N2O += R * (np.log(q_rot_rod(Θ_rot(B_tilde = 0.421019),T)) + f_rot_N2O/2)
    result_N2O += (S_bar_vib(Θ_vib(ν_tilde = 1276.5),T) + S_bar_vib(Θ_vib(ν_tilde = 589.2),T) * 2 + S_bar_vib(Θ_vib(ν_tilde = 2223.7),T))
    result_N2O += R * np.log(g_el_N2O)
    theoretical_array2[count] = result_N2O
    count += 1

# Chase, 1998 Shomate equation
A = 60.30274
B = 1.034566
C = -0.192997
D = 0.012540
E = -6.860254
F = 48.61390
G = 272.5002
H = 82.04824

count = 0
for T in T_array2:
    t = T / 1000
    value = A*np.log(t) + B*t + C*t**2/2 + D*t**3/3 - E /(2*t**2) + G
    shomate_array2[count] = value
    count += 1

plt.figure(figsize = (8,6))
plt.plot(T_array2, theoretical_array2, 'r-', lw = 4, label = "Theory")
plt.plot(T_array2, shomate_array2, 'b--', lw = 4, label = "Shomate equation, chase 1988")
plt.legend()
plt.xlabel("Temperature (K)", fontsize = 15)
plt.ylabel("Standard molar entropy of N2O $( J / (K mol))$", fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15)
```

```python
plt.figure(figsize = (8,6))
plt.semilogx(T_array, theoretical_array, 'r-', lw = 4, label = "Theory")
plt.semilogx(T_array, shomate_array, 'g--', lw = 4, label = "Shomate equation 1., chase 1988")
plt.semilogx(T_array2, theoretical_array2, 'r-', lw = 4)
plt.semilogx(T_array2, shomate_array2, 'b--', lw = 4, label = "Shomate equation 2., chase 1988")
plt.legend()
plt.xlim(100,10000)
plt.xlabel("Temperature (K)", fontsize = 15)
plt.ylabel("Standard molar entropy of N2O $( J / (K mol))$", fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.legend(fontsize = 15)
```

위의 두 결과를 확인해보시면 알겠지만, 저희가 푼 결과는 실험 결과를 매우 잘 설명하고 있습니다. (놀랍게도)

Shomate equation으로 Standard molar entropy의 값을 보여주는 함수는 모든 temperature range에서 적용되지 않아 lower temperature에 대한 식과, higher temperature에 대한 식이 다르나,

저희가 이론으로 구한 결과는 모든 temperature region에서 standard molar entropy를 예측하는 식이라는 것 또한 확인할 수 있으실 겁니다.

### 실험적으로 entropy는 어떻게 측정한 것일까요?

이전에 김지현 교수님께서 수업을 하신 내용을 되짚어보면 다음과 같은 식을 보신 기억이 있을 것입니다.



$$TdS = \delta q_{rev}$$

여기에서 $\delta q_{rev}$는 계(system)로 들어온 infinitesimal한 열입니다.

이 식을 통해 



$$\Delta S = \int_L \frac{\delta q_{rev}}{T}$$

가 성립하게 됩니다.

또한 [**"the third law of thermodynamics"**](https://www.wikiwand.com/en/Third_law_of_thermodynamics)에 의거, $T=0K$에서 $S=0$가 됩니다.

따라서, 절대 영도를 기준으로 매우 천천히(reversible한 process를 위해) 열($\delta q_{rev}$)을 넣어주면서, 이 열에 따라 바뀌는 온도($T$)의 결과를 계속 기록하고, 이러한 값들을 통해 entropy를 계산합니다.

이러한 것들을 들어봤을때, 실험적으로 entropy를 구하는 것은 많이 복잡하겠죠?

그러한 결과를 이론으로 예측가능하다는 것은, 특정 온도에 대한 entropy를 알기 위해 굳이 실험을 할 필요 없이 이론으로 값을 구할 수 있다는 것을 의미합니다.
