```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
length = 100

x = np.linspace(-10,10, length)
y = np.linspace(-10,10, length)

xx, yy = np.meshgrid(x,y) ## (xx, yy)로 구성된 (2 x length x length) tensor를 생성해주는 함수입니다.
                          ## 이를 통해, 보고자 하는 모든 point에서의 (xx, yy) 조합을 쉽게 얻을 수 있습니다.

z = xx**2 + yy**2
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

ax.plot_surface(xx, yy, z)
plt.show()
```
