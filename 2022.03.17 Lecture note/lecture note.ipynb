지난시간에 이어서 Quantum Mechanics에서 묘사하는 Angular momentum에 대해 수업을 진행했습니다.

Angular momentum은 위에 첨부한 그림에 설명된 것과 같이, 회전 축에 대한 particle의 위치 벡터($\textbf{r}$)와 particle이 가지는 운동량 벡터($\textbf{p}$) 사이의 cross product로 표현됩니다.

cross product를 "외적"이라고 표현한 책들이 많은데, outer product라는 cross product와 전혀 다른 operator가 있기에, 이 번역은 틀린 번역입니다. 벡터곱으로 표현하는 것이 맞습니다.

[**cross product**](https://www.wikiwand.com/ko/%EB%B2%A1%ED%84%B0%EA%B3%B1)를 통해 얻은 angular momentum은 다음과 같습니다.

처음 적힌 angular momentum은 고전 역학(classical mechanics)에서 정의된 angular momentum입니다. 이때는 $\textbf{r}$와 $\textbf{p}$ 모두 특정한 값으로 적을 수 있습니다.

하지만, 저희가 수업에서 angular momentum을 다루는 이유는, 분자의 rotational dynamics로 생기는 energy를 설명하기 위함입니다.

이 경우에는 고전 역학이 아닌 양자 역학(Quantum mechanics)적인 angular momentum이 고려되어야 됩니다.

이전 시간에 언급했듯이, 양자 역학에서는 particle의 momentum을 operator로 다룹니다.

이를 반영한 것이 뒤에 적인 양자역학에서의 angular momentum입니다.

이를 통해 angular momentum을 구성하는 각 성분들($\hat{L}_x$, $\hat{L}_y$, $\hat{L}_z$)에 대한 operator를 구했습니다.

Operator identities
===================

그렇다면, 위에서 구한 operator는 어떠한 성질을 가질까요?

이를 알아보기에 앞서, [**commutator**](https://quantummechanics.ucsd.edu/ph130a/130_notes/node109.html)에 대한 설명을 하고, 이를 활용하도록 하겠습니다.

$\psi$에 대해 두가지 operator $\hat{A}$와 $\hat{B}$가 있다고 할 때, operator $\hat{A}$와 $\hat{B}$에 대한 commutator는 다음과 같이 쓸 수 있습니다.

$$[\hat{A},\hat{B}]=\hat{A}\hat{B}-\hat{B}\hat{A}$$

이 commutator는 $\psi$에 다음과 같이 작용하게 됩니다.

$$[\hat{A},\hat{B}]\psi=\hat{A}\hat{B}\psi-\hat{B}\hat{A}\psi$$

operator는 $\psi$에 가까운 순으로 작용을 하게 됩니다.

만약에 $\hat{A}\psi = a\psi$이고, $\hat{B}\psi = b\psi$인 경우에, $\hat{A}\hat{B}\psi$는

$$\hat{A}\hat{B}\psi=\hat{A}b\psi=b\hat{A}\psi=ba\psi$$

가 됩니다. 

하지만 만약에 $\hat{A}\psi = a\psi$, $\hat{B}\psi = b\psi$, 그리고 $\hat{A}b = c$인 경우에, $\hat{A}\hat{B}\psi$는

$$\hat{A}\hat{B}\psi=\hat{A}b\psi=\hat{A}(b\psi)=(\hat{A}b)\psi+b(\hat{A}\psi)=c\psi+ba\psi=(c+ba)\psi$$

가 됩니다.

첫번째 경우에 $[\hat{A},\hat{B}]\psi$ 를 계산하면, $\hat{A}\hat{B}\psi=\hat{B}\hat{A}\psi=ab\psi$ 이므로 $[\hat{A},\hat{B}]\psi = 0$의 결과를 얻습니다.

그러나 두번째 경우에 대해 $[\hat{A},\hat{B}]\psi$를 계산하면, $\hat{B}\hat{A}\psi=ab\psi$ 이지만, $\hat{A}\hat{B}\psi=(c+ba)\psi$ 이기에, $[\hat{A},\hat{B}]\psi = c\psi$의 결과를 얻게 됩니다.

첫번째 경우에 대해서는 두 operator가 commute하다고 하고([**the two operators commute**](https://chem.libretexts.org/Courses/Pacific_Union_College/Quantum_Chemistry/04%3A_Postulates_and_Principles_of_Quantum_Mechanics/4.06%3A_Commuting_Operators_Allow_Infinite_Precision#:~:text=If%20two%20operators%20commute%2C%20then,B%3D%CB%86B%CB%86A.)), 두번째 경우에 대해서는 commute하지 않다고 합니다.



Commutator operator는 다음과 같은 성질을 가지고 있습니다.

1. $[\hat{B},\hat{A}]\psi = -[\hat{A},\hat{B}]\psi$

2. $[\hat{A}^2,\hat{B}]=\hat{A}[\hat{A},\hat{B}]+[\hat{A},\hat{B}]\hat{A}$

위에서 설명드린 commutator operator를 잘 생각해보면 이 성질이 당연하다는 것을 알 수 있을 것입니다.

-------------------------------------------------------------------------------------
이 commutator를 이용해서 다음과 같은 Angular momentum operator의 5가지 identy를 확인할 수 있습니다.

## 1. $[\hat{x},\hat{p}_x]\psi=i\hbar\psi$, $[\hat{y},\hat{p}_y]\psi=i\hbar\psi$, $[\hat{z},\hat{p}_z]\psi=i\hbar\psi$



## 2. $[\hat{L}_x,\hat{L}_y]\psi=i\hbar\hat{L}_z\psi$, $[\hat{L}_y,\hat{L}_z]\psi=i\hbar\hat{L}_x\psi$, $[\hat{L}_z,\hat{L}_x]\psi=i\hbar\hat{L}_y\psi$



## 3.$[\hat{L}_z, \hat{L}_\pm] = \pm \hbar \hat{L}_\pm$

### $\hat{L}_\pm \equiv \hat{L}_x \pm i \hat{L}_y$ : Ladder Operator (+: raising / -: lowering)



## 4. $[\hat{\textbf{L}}^2,\hat{L}_z]\psi=0$



## 5. $\hat{L}_{+}\hat{L}_{-}=(\hat{L}_x+i\hat{L}_y)(\hat{L}_x-i\hat{L}_y)=\hat{L}_x^2+\hat{L}_y^2-i[\hat{L}_x,\hat{L}_y]=\hat{L}_x^2+\hat{L}_y^2+\hbar\hat{L}_z=\hat{L}^2-\hat{L}_z^2+\hbar\hat{L}_z$

---------------------------------------------------------------

이제 위에서 얻은 5가지 properties를 이용해서 Angular momentum에 대해 다뤄보자.




우선 들어가기에 앞서, $\hat{L}_z\varphi_m =\hbar m\varphi_m$이라는 식을 깔고 시작한다.

즉, $\hbar m$은 $\hat{L}_z$ operator에 대한 eigenvalue이다.

이에 대한 물리적 의미는 위와 같이 생각할 수 있다.

위에서 $2\pi r$에 대해서 $2\pi r=m\lambda$이 맞을지, $2\pi r=m\frac{\lambda}{2}$가 맞는지 고민을 해볼 수 있다.

하지만 끝과 끝이 만나는 지점에서 미분값이 존재하기 위해서는 $2\pi r=m\lambda$로 보는 것이 맞다.

### 1. Ladder Operator 활용

$$\hat{L}_z \varphi_m = \hbar m \varphi_m$$

$$\hat{L}_z \hat{L}_+ \varphi_m = ([\hat{L}_z,\hat{L}_+] + \hat{L}_+ \hat{L}_z)\varphi_m = \hbar (m+1) (\hat{L}_+ \varphi_m)$$

$$\hat{L}_z \hat{L}_- \varphi_m = ([\hat{L}_z,\hat{L}_-] + \hat{L}_- \hat{L}_z)\varphi_m = \hbar (m-1) (\hat{L}_- \varphi_m)$$

위와 같이, $\hat{L}_\pm \varphi_m$은, $\varphi_m$와 비교했을시, $\hat{L}_z$ operator에 대한 eigenvalue가 1만큼 증가하거나 감소한다.

이러한 성질을 가졌기 때문에 이 operator를 ladder operator라고 한다. ladder operator는 $\varphi_m$의 $L_z$ 값을 증감시키는 operator라고 볼 수 있다.

### 2. m value에 한계가 있는가?

Total angular momentum($\textbf{L}^2$) conservation을 고려한다면, $L_z$ 값에는 한계가 있을 수 밖에 없다.

$\hat{\textbf{L}}^2$의 eigenvalue를 $\hbar^2 k^2$ 라 한다면, $m$은 $m^2 \leq k^2$을 만족해야 된다.

$\because \hat{\textbf{L}}^2 = \hat{L}_x^2 + \hat{L}_y^2 + \hat{L}_z^2$ 이므로 $\hat{L}_z^2 \leq \hat{\textbf{L}}^2$

$m$이 가질 수 있는 최소 값을 $m_{min}$이라 하고, 최대 값을 $m_{max}$라 하면, 

$\varphi_{m_{max}}$에 $\hat{L}_+$ operator가 작용하게되면 $\hat{L}_+ \varphi_{m_{max}} = 0$이 되고,

또한 $\varphi_{m_{min}}$에 $\hat{L}_-$ operator가 작용하게되면 $\hat{L}_- \varphi_{m_{min}} = 0$이 된다.
