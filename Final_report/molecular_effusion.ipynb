기체 분자가 저장되어있는 container를 상상해봅시다. 

이 container에 작은 구멍이 나는 경우, 풍선에서 바람이 빠지는 것처럼, 기체 분자들이 바깥으로 나오게 될 것입니다.

이렇게 기체 분자가 나오는 것은, 완전히 container가 열려있을 때와 다른 상황입니다.

container가 완전히 열려있다면, 기체 분자는 확산(diffusion)을 통해 container에서 나오게 되겠지만,

container에 작은 구멍이 났다면, 기체 분자는 container가 가지는 내부 압력이 강하면 강할수록 빠르게 나오는 상황을 겪게 됩니다.

이러한 현상을 molecular effusion이라고 합니다.

이 molecular effusion의 flux는 단위시간동안 단위 면적의 구멍을 통해 나오는 분자의 개수로 정의가 됩니다.

수업시간에 frequency of collsions $Z_{coll}$이라는 개념을 배운 바 있습니다.

이 개념을 잘 숙지하셨다면, effusion flux는 쉽게 이해하실 수 있으실 것입니다.

frequency of collision이 단위 시간동안 단위 면적당 벽에 가해지는 기체 분자의 collision 횟수인데,

이 벽에 구멍이 뚤려있다면..? 기체 분자는 벽과 충돌을 하지 못하고 container를 나가게 됩니다.

즉, $Z_{coll}$의 식이 그대로 effusion flux에 사용이 가능하다는 의미입니다.

이제 이 식을 정리해봅시다.

$$\rho = \frac{N}{V} = \frac{P}{k_B T}$$

$$\braket{u} = \sqrt{\frac{8 k_B T}{\pi m}}$$

이 두가지 식을 이용해 정리를 하게 되면,

$$Z_{effuse} = \rho \frac{\braket{u}}{4} = \frac{P}{k_B T} \times \frac{1}{4} \times \sqrt{\frac{8 k_B T}{\pi m}} = \frac{P}{\sqrt{2 \pi m k_B T}} $$

이는 문제에 주어진 effusion flux의 식과 같게 됩니다.

-------------------------------------

그렇다면, 문제에서 주어진 조건에 대해 단위 시간당 molecular effusion을 통해 container를 나가는 질소의 개수는 어떻게 계산을 해야 될까요?

이는 간단하게,

$$Z_{effuse} = \frac{1}{A} \frac{d N_{effuse}}{dt}$$

라는 조건을 이용해 $Z_{effuse} \times A$를 계산하면 됩니다.

```python
import numpy as np

# constant values
N_avo = 6.022 * 10**23 # [1/mol]
k_B = 1.38 * 10**-23 # [J/K = N m / K = kg m/s**2 m/K = kg (m/s)**2 / K]
P = 1 * 10**5 # 1 bar [N m**-2 = kg * m**-1 * s**-2]
m_N2 = 28.014 # Nitrogen (N2) #[g/mol]
T = 25 + 273.15 #[K]
r = 0.01/2 * 10**-3 #[m]


# mass calculation function
def mass(M_r): # molecular weight to mass per particle

    # 분자량(M_r)을 넣어주면, 이 분자 하나 당 질량을 계산해 줍니다.
    return M_r / N_avo * 10**-3 # 단위: [kg]


Z_effuse = P / np.sqrt(2 * np.pi * mass(m_N2) * k_B * T) 
# sqrt(= square root) 내부의 단위는, kg * J = (kg m/s)**2 이다.
# 그렇기에, Z_effuse의 단위는 [kg * m**-1 * s**-2 / (kg * m * s**-1) = m**-2 * s**-1] 이 된다.

Area_of_hole = np.pi * r**2 #[m**2]

result = Z_effuse * Area_of_hole # [number/s]

print(f'매 초마다 {result} 개의 질소 분자가 반지름이 {r * 10**3}[mm] 인 구멍을 통해 container에서 나간다.')
```

-----------------------------------

위에서 얻은 결과를 다시 한번 살펴봅시다.

분자의 effusion flux는 다음과 같이 정리됩니다.

$$Z_{effuse}=\frac{P}{\sqrt{2 \pi m k_B T}}$$

이를 통해 알 수 있는 사실은,

1. 압력이 높을수록 effusion되는 분자의 개수가 증가하게 됩니다. ($\because Z_{effuse} \sim P$)

2. effusion이 발생할 때, 무거운 분자는 가벼운 분자에 비해 container를 탈출하지 못합니다. ($\because Z_{effuse} \sim m^{-1/2}$)

이러한 성질을 활용하면, effusion을 이용해 무거운 분자와 가벼운 분자를 어느정도 분리하는 것이 가능하다는 것을 생각해볼 수 있습니다.

역사적으로도 이 effusion이 분자를 분리하는데 사용된 적이 있습니다.
