```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = 100
y = 50
x = np.linspace(0,L,10000)
```

```python
def psi_n(x,n,L):
    return np.sqrt(2/L)*np.sin(n/L*np.pi*x)
```

```python
def dirac_delta(x,y,n,L):
    result = 0
    for m in range(1,n+1):
        result += psi_n(x,m,L)*psi_n(y,m,L)
    return result
```

```python
value = dirac_delta(x,y,1,L)
plt.plot(x,value**2)
```

```python
value = dirac_delta(x,y,100000,L)
plt.plot(x,value**2)
```

```python
# number = 100000
number = 10**5

plt.plot(x,value**2)
plt.text(0,0.005,f'n = {number}',fontsize=20)
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
filenames = []

for m in range(1,15):
    plt.plot(x,dirac_delta(x,y,2**m,L)**2)
    plt.plot(figsize=(20,20))
    number = m
    plt.text(0, 0, f'n = {number}', fontsize = 20)
    filename = f'{m}.png' # f-string
    filenames.append(filename)
    
    plt.savefig(filename)
    plt.close()
    
with imageio.get_writer('result.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
        writer.append_data(image)

for filename in set(filenames):
    os.remove(filename)
```
